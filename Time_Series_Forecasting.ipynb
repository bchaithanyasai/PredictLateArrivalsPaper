```
import datetime
import numpy as np

def format_heure(time):
    if pd.isnull(time):
        return np.nan
    else:
        if time == 2400: time = 0
        time = "{0:04d}".format(int(time))
        heure = datetime.time(int(time[0:2]), int(time[2:4]))
        return heure

df['converted_DEPARTURE_TIME'] = df['DEPARTURE_TIME'].apply(format_heure)
df
```
